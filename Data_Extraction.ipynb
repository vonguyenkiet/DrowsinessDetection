```python
import os
import shutil
from moviepy import VideoFileClip

# ================= CẤU HÌNH ĐƯỜNG DẪN =================
INPUT_DIR = r"C:\Users\Kiet\Downloads\68cs2\học kì 6\Thị giác máy tính\Fold1_part1" 
OUTPUT_DIR = r"C:\Users\Kiet\Downloads\68cs2\học kì 6\Thị giác máy tính\DMS_Project\data\raw" 

# ================= TỰ ĐỘNG GÁN NHÃN ===================
FILE_TO_LABEL_MAP = {
    "0": "Normal",       
    "5": "Distracted",   
    "10": "Drowsy"       
}

def setup_directories():
    for label in FILE_TO_LABEL_MAP.values():
        os.makedirs(os.path.join(OUTPUT_DIR, label), exist_ok=True)

def auto_extract_and_label():
    setup_directories()
    
    # Quét 12 người
    for person_id in range(1, 13):
        folder_name = f"{person_id:02d}"
        person_folder = os.path.join(INPUT_DIR, folder_name)
        if not os.path.isdir(person_folder):
            person_folder = os.path.join(INPUT_DIR, str(person_id))
            if not os.path.isdir(person_folder):
                continue
            
        videos = [f for f in os.listdir(person_folder) if f.lower().endswith(('.mp4', '.avi', '.mov'))]
        
        for video_idx, video_name in enumerate(videos, start=1):
            video_path = os.path.join(person_folder, video_name)
            file_prefix = os.path.splitext(video_name)[0]
            
            if file_prefix not in FILE_TO_LABEL_MAP:
                print(f"Bỏ qua file {video_name} vì không có trong Map.")
                continue
                
            auto_label = FILE_TO_LABEL_MAP[file_prefix]
            print(f"\n--- Đang Auto-Cắt: Người {person_id} - {video_name} => Nhãn: [{auto_label}] ---")
            
            try:
                clip = VideoFileClip(video_path)
                duration = int(clip.duration)
                
                for start_t in range(0, duration, 20):
                    end_t = min(start_t + 20, duration)
                    if end_t - start_t < 2: 
                        continue 
                    
                    final_filename = f"{auto_label}_{person_id}_{file_prefix}_{start_t}.mp4"
                    final_path = os.path.join(OUTPUT_DIR, auto_label, final_filename)
                    
                    print(f" -> Lưu file: {final_filename}")
                    
                    # Đã fix lỗi subclipped cho bản MoviePy v2
                    subclip = clip.subclipped(start_t, end_t)
                    subclip.write_videofile(final_path, codec="libx264", audio=False, logger=None)
                    
                # ĐÓNG FILE TRƯỚC KHI XOÁ
                clip.close()
                
                # XOÁ FILE THÔ ĐỂ GIẢI PHÓNG DUNG LƯỢNG
                os.remove(video_path)
                print(f" [!] Đã xoá file gốc để giải phóng dung lượng: {video_name}")
                
            except Exception as e:
                print(f" [LỖI] Xảy ra lỗi với file {video_name}: {str(e)}")

    print("\n[+] HOÀN THÀNH TỰ ĐỘNG CẮT, GÁN NHÃN VÀ DỌN DẸP DỮ LIỆU THÔ!")

if __name__ == "__main__":
    auto_extract_and_label()
```
